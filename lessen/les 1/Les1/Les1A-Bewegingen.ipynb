<h2>Inhoud van het deel &ldquo;Beweging&rdquo;</h2>
<h3>Technische reikwijdte en kennis</h3>
<ul>
<li><b>kinematica</b>: geometrische beweging over de tijd, van een mechanisch gekoppeld systeem van
<a href="https://en.wikipedia.org/wiki/Rigid_body_dynamics"><b>starre lichamen</b></a>.
Dat wil zeggen, je kan de <em>positie</em>, <em>snelheid</em>, en <em>versnelling</em> berekenen van gelijk welk punt op gelijk welk star lichaam in het systeem.
<li><b>dynamica</b>: het effect dat <b>krachten</b> hebben op de beweging van het systeem.
Als het starre lichaam in rust is en blijft, dan noemen we dat de
(<a href="https://en.wikipedia.org/wiki/Statics"><b>statica</b></a>),
of &ldquo;krachtenevenwicht&rdquo;.
<br>Als de krachten de beweging van de starre lichamen wijzigt, dan noemen we dat de
<a href="https://en.wikipedia.org/wiki/Rigid_body_dynamics"><b>kinetica</b></a>.
De onderliggende fysische wetmatigheden zijn de
<a href="https://en.wikipedia.org/wiki/Newton%27s_laws_of_motion">wetten van Newton</a>, 
via de
<a href="https://en.wikipedia.org/wiki/List_of_moments_of_inertia"><b>inertie</b></a>
van de starre lichamen.

<li><b>(stangen)-mechanismes</b> (zoals het
<a href="https://en.wikipedia.org/wiki/Slider-crank_linkage">kruk-drijfstang</a>-mechanisme in een
<a href="https://en.wikipedia.org/wiki/Internal_combustion_engine">verbrandingsmotor</a>),
en <b>nok-aandrijvingen</b>:
<ul>
<li>de <b>mechanische koppeling</b> zorgt voor een <b>deterministische</b> relatieve beweging.
<li>in deze cursus beschouwen we enkel systemen met
<a href="https://en.wikipedia.org/wiki/Degrees_of_freedom_(mechanics)"><b>&eacute;&eacute;n vrijheidsgraad</b></a>.
Met andere woorden, &eacute;&eacute;n enkele
<a href="https://en.wikipedia.org/wiki/Actuator">actuator</a> (of &ldquo;motor&rdquo;) is voldoende voor de aandrijving van het systeem.
</ul>

<li><b> machines</b> (zoals de hogervermelde verbrandingsmotor), zijn een collectie van mechanismes, aangedreven door een collectie van actuatoren, om een nuttige beweging te genereren voor de maatschappij.
</ul>

We bestuderen de <b>traject-berekeningen</b> van mechanismes en nokken, en
de <b>verbanden</b> met de <b>trillingen</b> die mogelijkerwijze zouden kunnen ontstaan als een gevolg van die bewegingen.
