<h2>Beweging en Trillingen</h2>
<p>Het vak <a href="https://onderwijsaanbod.kuleuven.be/syllabi/n/H01N0AN.htm#activetab=doelstellingen_idp1557904">Beweging en trillingen</a> 
bestaat uit twee delen:
<ul>
<li><b>Deel I: Kinematica en dynamica van machines</b> (&ldquo;beweging&rdquo;, &ldquo;motion&rdquo;)
<br>
Herman Bruyninckx. Alexander Badri-Spröwitz
<p>Dit deel heeft twee sub-delen:
<ul>
<li><a href="https://en.wikipedia.org/wiki/Mechanism_(engineering)">mechanismes</a>:
<b>starre lichamen</b> verbonden met <b>rotatie</b>- en/of <b>translatie</b>-gewrichten.
<div style="height: 10rem;">
 <img src="./images/Hexapod_general_Anim.gif" height="100%" />
 <img src="./images/Pantograph_Mirror.gif" height="100%" />
</div>
<li><b>nokken</b>: <b>starre</b> lichamen verbonden door een &ldquo;rollend contact&rdquo; tussen beide.
<div style="height: 10rem;">
 <img src="./images/220px-Nockenwelle.gif" height="100%" />
 <img src="./images/grabcad-dual-sided-cam.gif" height="100%" />
</div>
</ul>
<p>In beide delen is de <em>dynamica</em> beperkt tot alleen de effecten van <b>inertie</b>
(&ldquo;<a href="https://en.wikipedia.org/wiki/Newton's_laws_of_motion">wet van Newton</a>&rdquo;).
De nadruk ligt op machines die <b>repetitieve bewegingen</b> uitvoeren met een vaste cyclustijd en (nagenoeg) constante snelheid.

<li><b>Deel II: Mechanische trillingen</b> (&ldquo;trillingen&rdquo;, &ldquo;vibrations&rdquo;)
<br>
Elke Deckers.
<br>
De <em>dynamica</em> bevat ook de effecten van <b>elasticiteit</b>
(&ldquo;<a href="https://en.wikipedia.org/wiki/Hooke%27s_law">wet van Hooke</a>&rdquo;)
zodat trillingen kunnen optreden als uitwisseling van kinetische energie (opgeslagen in de bewegende inertie) en potentiele energie (opgeslagen in de elasticiteit).
</ul>

<h2>Evaluatie van het gedeelte &ldquo;Beweging&rdquo;</h2>
<p>Gebeurt aan de hand van <b>twee projecten</b>:
<ul>
<li>&eacute;&eacute;n over <b>stangenmechanismen</b>, en &eacute;&eacute;n over <b>nokken</b>.
<li>elk project gebeurt door een <b>team van twee studenten</b>.
<li>elk project is een <em>Jupyter notebook</em>, zoals dit document. 
Je levert die in aan het einde van het semester.
<li>zo'n <em>notebook</em> bevat alle berekeningen, dus kunnen we makkelijk &ldquo;wat als?&rdquo; vragen stellen op het <b>mondelinge examen</b>.
<li>er is een <b>gezamenlijk mondeling examen</b> van 20 minuten, voor beide projecten.
</ul>



<p><b>Tijdsbudget/student</b> (totaal = 90 uren):
<ul>
<li>lessen: 14 uur.
<li>twee (of drie) gezamenlijke oefensessies: 4 uur.
<li>projecten: &approx; 30 + 30 = 60 uur.
</ul>
<p><b>Score-criteria</b>:
<ul>
<li><span style="color: red;"><b>inzichten</b></span>. Worden getest op basis van &ldquo;waarom?&rdquo; en &ldquo;wat als?&rdquo; vragen.
<li><b>vier hoofdvragen</b> te beantwoorden in 20 minuten. Elke vraag die niet aan bod kan komen resulteert in een verlies van 2.5/10.
<li>correcte berekeningen.
<li>creativiteit en kritische reflectie. Bijvoorbeeld: welke industrie is gebaat met jullie stangenmechanisme, en jullie nokken?
</ul>
