<p>Waarom vliegwiel?</p>
<ul>
<li><b>Stabiliseren van de draaisnelheid:</b> In verbrandingsmotoren levert de ontbranding schokkerige krachten. Het vliegwiel vlakt deze "stoten" af, waardoor de motor trillingsvrij en gelijkmatig blijft draaien.
<li><b>Energieopslag:</b> Het vliegwiel slaat kinetische energie op tijdens momenten van overvloed en geeft deze af wanneer er extra kracht nodig is. Dit wordt bijvoorbeeld toegepast bij energieopslagsystemen die elektriciteitsnetten in balans houden.
<li><b>Hulp bij dode punten:</b> In mechanismen zoals een stoommachine helpt het vliegwiel de beweging door de 'dode punten' (momenten zonder aandrijving) heen te drukken, zodat de rotatie niet stopt.
<li><b>Overwinnen van traagheid:</b> Het helpt bij het opstarten van motoren door voldoende impuls te behouden om de eerste verbrandingsslagen soepel te laten verlopen.
<li><b>Gyroscopische stabiliteit:</b> In instrumenten of voertuigen zoals satellieten worden vliegwielen (reactiewielen) gebruikt om de oriëntatie stabiel te houden. 
</ul>
<p>Nadelen van vliegwiel?</p>
<ul>
<li><b>Gewicht en omvang:</b> Voor een effectieve energieopslag moet een vliegwiel vaak zwaar of erg groot zijn. Dit maakt ze minder geschikt voor compacte, mobiele toepassingen waar elke kilo telt.
<li><b>Veiligheidsrisico's:</b> Door de extreem hoge rotatiesnelheden (soms tienduizenden toeren per minuut) kan een vliegwiel bij een defect letterlijk uit elkaar spatten. Dit vereist zeer sterke en dure insluitingsbehuizingen om schade door rondvliegende brokstukken te voorkomen.
<li><b>Mechanische slijtage:</b> De lagers worden zwaar belast door de constante rotatie en hoge krachten, wat kan leiden tot lagerfalen en regelmatige onderhoudskosten.
<li><b>Beperkte opslagduur:</b> Vliegwielen verliezen relatief snel energie door wrijving (zelfs in een vacuüm), waardoor ze vooral geschikt zijn voor kortstondige opslag en minder voor het dagenlang bewaren van energie.
<li><b>Gyroscopisch effect:</b> In bewegende voertuigen kan de draaiing van het vliegwiel de besturing beïnvloeden. Het vliegwiel wil namelijk in dezelfde stand blijven staan, wat tegenkrachten opwekt bij het nemen van bochten.
<li><b>Hogere kosten:</b> De noodzaak voor hoogwaardige materialen (zoals koolstofvezel), precisie-balancering en speciale behuizingen maakt de initiële investering vaak hoger dan bij chemische batterijen. 
</ul>

<h3>Mobiliteit in het vlak</h3>

<p>(Synomiemen voor &ldquo;mobiliteit&rdquo;: &ldquo;beweegbaarheid&rdquo;, &ldquo;aantal vrijheidsgraden&rdquo;)</p>

<p>De <a href="https://en.wikipedia.org/wiki/Linkage_(mechanical)">mobiliteit</a> van een mechanisme is zijn aantal <em>vrijheidsgraden</em> (degrees of freedom, &ldquo;DOF&rdquo;). 
Met andere woorden: het <em>minimum</em> aantal <em>onafhankelijke</em> coordinaten nodig om de positie en orientatie voor te stellen van <em>elk</em> gelid op <em>elk</em> ogenblik in de tijd.

<p>Mobiliteit in het vlak: formule van 
<a href="https://en.wikipedia.org/wiki/Chebychev%E2%80%93Gr%C3%BCbler%E2%80%93Kutzbach_criterion">Chebychev–Grübler–Kutzbach</a>:

$$M = {\color{red}3}(n − 1) − {\color{blue}2}f_1 − f_2$$


$n$: aantal <em>gelederen</em>.
Elke vast gelid verwijdert <span style="color: red;">3</span> DOFs.
<br>
$f_1$: aantal <em>gewrichten</em> met 1 DOF (rotatie of prismatisch).
Elk verwijdert <span style="color: blue;">2</span> DOFs.
<br>
$f_2$: aantal <em>gewrichten</em> met 2 DOFs (nok, volgers, rollers),
Elk verwijdert <em>1</em> DOFs.

<h3>Mobiliteit in de 3D ruimte</h3>
Een zelfde redenering als in het 2D vlak leidt tot de volgende 3D formulering van de 
<a href="https://en.wikipedia.org/wiki/Chebychev%E2%80%93Gr%C3%BCbler%E2%80%93Kutzbach_criterion">Chebychev–Grübler–Kutzbach formule</a>:

$$M = \color{red}{6}(n − 1) − \color{blue}{5} f_1 − \color{green}{4} f_2− \color{Maroon}{3} f_3  − \color{DarkGreen}{2} f_4 - \color{DarkBlue}{1} f_5$$

<p>

$n$: aantal <em>gelederen</em>. Elk vast gelid verwijdert <span style="color: red;">6</span> DOFs.
<br> $f_1$: aantal 1 DOF <em>gewrichten</em>, elk verwijdert <span style="color: blue;">5</span> DOF.
Bijv.: <em>helisch/schroef</em>-gewricht.
<br> $f_2$: aantal 2 DOF <em>gewrichten</em>, elk verwijdert <span style="color: green;">4</span> DOF.
Bijv.: <em>cylindrisch</em> -gewricht.
<br> $f_3$: aantal 3 DOF <em>gewrichten</em>, elk verwijdert <span style="color: Maroon;">3</span> DOF.
Bijv.: <em>bal-/sferisch</em> gewricht.
<br> $f_4$: aantal 4 DOF <em>gewrichten</em>, elk verwijdert <span style="color: DarkGreen;">2</span> DOF.
Bijv.: ??
<br> $f_5$: aantal 5 DOF <em>gewrichten</em>, elk verwijdert <span style="color: DarkBlue;">1</span> DOF.
Bijv.: ??
</p>

<h3>Mobiliteit in 3D ruimte versus in 2D ruimte</h3>

<div style="display: grid; grid-template-columns: 3fr 2fr; grid-column-gap: 5%;">
<div>

$n$ = 4; $f_1$ = 4.

<b>2D</b>: $M = 3(n − 1) − 2 f_1 = 1$

<b>3D</b>: $M = 6(n − 1) − 5f_1 = −2$?

<p><b>Verklaring</b>: in het vlak geldt een <b>extra, impliciete <em>geometrische</em></b> bewegingsbeperking, met name dat elk gelid enkel in <em>hetzelfde vlak</em> mag bewegen.

&rarr; alle gewrichtsassen zijn <b>parallel</b>.

&rarr; dit neemt <em>altijd</em> 3 DOFs weg per gelid.

(Tenminste, als de <em>mechanische uitlijning</em> perfect is!)
