<h3>Bewegingswetten</h3>

<p><b>Doelstellingen van deze les</b>.
<ul>
<li>een bewegingswet is een wiskundige beschrijving van de beweging van <b>een punt</b>. In de meeste gevallen is dat een punt op een star lichaam waarvan de beweging relevant is voor een bepaalde <b>toepassing</b>.
<li>bewegingswetten zijn algemeen toepasbaar: individuele lichamen, mechanismes, nokken,&hellip;
<li>we introduceren de meest voorkomende bewegingswetten, in een <b>dimensieloze</b> representatie, die analyse, ontwerp, en vergelijking tussen toepassingen, vereenvoudigt.
</ul>

</div>

<div style="width: 76ch;">

<h3>Relevantie voor mechanismes en nokken</h3>

<ul>
<li><b>mechanisme</b>: door het hoog aantal sluitingsbeperkingen kan een gewenste bewegingswet vaak niet volledig gerealiseerd worden, en <em>benaderingen</em> zijn nodig.
<li><b>nok/volger-systemen</b> en <b>servo-aandrijvingen</b> (zoals robots of (&ldquo;<a href="https://en.wikipedia.org/wiki/Numerical_control">CNC</a>&rdquo;) werktuigmachines): deze hebben meestal meer bewegingsvrijheid, dus &ldquo;perfect&rdquo; volgen van een bewegingswet is mogelijk, tenminste toch in een <em>kinematische</em> analyse of ontwerp.
</ul>
De <em>dynamische</em> aspecten van nokken en servo-aandrijvingen zorgen bijna altijd <em>wel</em> voor afwijkingen, omdat de motoren altijd slechts een <em>eindige krachtcapaciteit</em> hebben.

In het geval van <em>nokken</em> noemt men de bewegingswet vaak ook de <b>heffingswet</b> van de nok, omdat de beweging zorgt voor de &ldquo;heffing&rdquo; van de volger t.o.v. zijn ruststand. Bij <em>servo-aandrijvingen</em> spreekt men over een <b>(bewegings-)traject</b>.


<h3>Nok/volger systeem</h3>

<div style="display: grid; grid-template-columns: 1fr 1fr; grid-column-gap: 1%;">
<div>

De <b>nok</b> roteert (in de meeste toepassingen) aan een <em>constante</em> hoeksnelheid $\omega$, dus de hoek $\theta = \omega \; t$. Tenminste, dat is meestal de bedoeling, maar door constructie- en/of dynamische onvolmaaktheden zijn er afwijkingen.

De <b>volger</b> volgt de nok, en heeft meestal een erg niet-constante op- en neergaande beweging $x$ als functie van $\theta$, als een gevolg van het **profiel** van de nok.
Die relatie **$x(\theta)$** noemen we de **bewegingswet**, of **heffingswet**.

</div>

Zoals bij mechanismes is de beweging van de volger <b>helemaal bepaald</b> door (1) het <b>mechanische profiel</b> van de nok, en (2) de <b>rotatie-beweging</b> van de nok. Als we dus een andere bewegingswet willen, dan moeten we de mechanische hardware veranderen. Toch worden nokken nog heel veel gebruikt, in toepassingen met snelle, repetitieve bewegingen.

<h3>Servo-aandrijvingen</h3>

De beweging van de last wordt gegenereerd door <em>computergestuurde motoren</em>, en dus kan de <b>gewenste beweging</b> $x(t)$ (of $x(k\; T_s)$ met $T_s$ de </em>bemonsteringstijd</em> van de computersturing) veranderd worden door <em>herprogrammeren</em> van de <b>software</b>, en niet (alleen) door verandering in de mechanische <b>hardware</b>.

Deze flexibiliteit is dus het relatieve voordeel van servo-aandrijvingen t.o.v. nok/volger-systemen. Het relatieve nadeel is de nood aan meer en krachtigere motoren, en meer aansturingssoftware.
</div>

<div style="width: 76ch;">

<h3>Meest voorkomende types van bewegingswetten</h3>

1. <b>Dubbele stilstand</b> beweging.
Er zijn twee momenten (of langere periodes) van stilstand (&ldquo;dwell&rdquo;), waar snelheid nul is, net als alle andere afgeleiden van positie.

2. <b>Periodische</b> beweging.
De beweging herhaalt zich na een bepaalde periode in de tijd, en daar zijn positie en al zijn afgeleiden gelijk in elke periode.

Een bewegingswet kan bestaan uit <em>&eacute;&eacute;n enkele functie</em> over de hele periode.
Een andere aanpak gebruikt <em>stuksgewijze functies</em> (waaronder <em>stilstands</em>-functies). 
Heel vaak gebruikt men voor de eerste <b><a href="https://nl.wikipedia.org/wiki/Polynoom">polynomen</a></b>, en voor de tweede <b><a href="https://nl.wikipedia.org/wiki/Spline">splines</a></b>. En de combinatie van <em>polynomiale splines</em> is vanzelfsprekend populair.

<h3>Orde van continu&iuml;teit (&ldquo;smoothness&rdquo;)</h3>
Hoe minder continu een beweging is in positie, snelheid, versnelling,&hellip;, hoe meer die beweging de <em>eigenfrekwenties</em> van het bewegende systeem kan aanslaan. (Zie deel &ldquo;Trillingen&rdquo; van dit vak.)

Een beweging heeft $n$-<em>de orde</em> continu&iuml;teit als 
- alle afgeleiden tot en met orde $n-1$ continu zijn, gedurende de hele beweging.
- de $n$-de orde afgeleide heeft &eacute;&eacute;n of meer discontinu&iuml;teiten.
- (dus) de $n+1$-ste afgeleide wordt (althans wiskundig gesproken) &ldquo;oneindig&rdquo; op &eacute;&eacute;n of meer plaatsen.

In het geval van polynomiale bewegingswetten bereikt men continu&iuml;teit van orde $n$ enkel met een polynoom van orde $2n$, omdat de continu&iuml;teit moet voldaan zijn aan <em>beide</em> einden van de beweging.

</div>

<div style="width: 76ch;">

<h3 id="fundWet">&ldquo;Fundamentele wet&rdquo; van nokken-ontwerp</h3>

_"De bewegingswet $x(\theta)$ moet continu zijn in de eerste en tweede afgeleiden van de verplaatsing van de nok, gedurende een hele periode $0 \le \theta \le 2\pi$."_

Dit is niet echt een <em>fysische wet</em>, maar eerder een <em>pragmatische richtlijn</em>. Het is een afweging (&ldquo;trade-off&rdquo;) tussen zachtheid van een bewegingsprofiel, en de productie-kosten van de nok: hogere garanties over de zachtheid van een mechanisch profiel vereist hogere nauwkeurigheid van de werktuigmachine die het profiel maakt.

Een gevolg van de fundamentele wet is dat de <b>ruk</b> (of &ldquo;jerk&rdquo;, met name de <em>afgeleide van de versnelling</em>) eindig is gedurende de hele beweging.
De fysische interpretatie van <em>versnelling</em> volgt uit de wet van Newton: versnelling is evenredig met de <em>inerti&euml;le belasting</em> van het systeem. Op het niveau van een tijdsafgeleide hoger interpreteren we de <em>ruk</em> dus als evenredig met de <em>verandering van de inerti&euml;le krachten</em>.


</div>

<h3>Dimensieloze voorstelling van dubbele-stilstand-beweging</h3>

<b>Probleemstelling</b>:

- <b>servo-aandrijving</b>: het <em>traject</em> van $0$ tot $L$ is
$$S(t) \quad (0 \le t \le T) : \quad S(0) =0 \text{ en } S(T) = L,$$
met $L$ de volledige af te leggen weg in translatie of rotatie.

- <b>nok/volger-systeem</b>: de <em>heffingswet</em> over een hoek $\beta = \omega \, t$ van de periode is
$$S(t) \quad (0 \le t \le T) : \quad S(0) =0 \text{ en } S(\beta) = L,$$
met $L$ de totale <b>heffing</b> (&ldquo;lift&rdquo;) in translatie of rotatie, af te leggen gedurende de rotatie van de nok over de hoek $\beta$.


De <b>dimensieloze voorstelling</b> normaliseert de tijd-co&ouml;rdinaat van de beweging naar de totale duur van de beweging, en de afgelegde weg-co&ouml;rdinaat naar de totale afgelegde weg:
$$s(\tau) \quad (0\le \tau \le 1): \quad s(0) =0, s(1) =1.$$

\begin{aligned}
\text{servo :}\quad & S(t) = L\; s(\tau), \tau = \frac{t}{T}.
\\
\text{nok/volger:} \quad & S(\theta =\omega \,t) = L\; s(\tau), \tau = \frac{\theta}{\beta}.
\end{aligned}

De <b>dimensieloze voorstelling van de afgeleiden</b> is voor de hand liggend:

\begin{aligned}
\text{servo:}\quad & \frac{dS(t)}{dt} = L\; s'(\tau)\; \frac{d\tau}{dt} = \frac{L}{T} \; s'(\tau),
\\
\text{en}\quad & \frac{dS^n(t)}{dt} = \frac{L}{T^n} \; s^{(n)}(\tau).
\\
\text{nok/volger:}\quad  &  \frac{dS(t)}{dt} = L\; s'(\tau)\; \frac{d\tau}{dt} = \frac{L}{\beta} \; \omega\; s'(\tau),
\\
\text{en}\quad & \frac{dS^n(t)}{dt} = \frac{L}{\beta^n} \; \omega^n \; s^{(n)}(\tau).
\end{aligned}

<b>Symbolen</b>:

\begin{aligned}
\text{snelheid:}    & \quad & v =s'(\tau), \\
\text{versnelling:} & & a = s''(\tau), \\
\text{ruk:}& & j = s^{(3)} (\tau), \\
\text{("jerk")}& & \\
\text{afgeleide van de ruk:} & & dj = s^{(4)} (\tau). \\
\text{("djerk", "ping")} & & 
\end{aligned}

<h3>Eigenschappen van dubbele-stilstand-bewegingen</h3>

De beschrijvingen zijn in <em>dimensieloze representatie</em>.
De volgorde van beschrijving is volgens de <em>orde van continu&iuml;teit</em>. Voor elke bewegings-functie geven we voor- en nadelen.

<b>Constante snelheid</b>

- <em>bewegingswet</em>: $$ s(\tau) = \tau.$$

<div style=" display: grid; place-items: center; place-content: center;">
 <img style="border: 0; display: inline-block;" height="100%"
      src="images/motion-law-constant-speed.png"
 >
</div>

- <em>continu&iuml;teit</em>: 1ste orde. Versnelling wordt oneindig bij $\tau=0$ en $\tau=1$.
- <em>voordeel</em>: deze bewegingswet heeft de <b>laagste piek-snelheid</b>, en een heel eenvoudige aansturing op de motor.
- <em>nadeel</em>: het vragen van een &ldquo;oneindige&rdquo; versnelling resulteert in hoge dynamische belasting (van motor <em>en</em> machine).


<b>Bang-bang versnelling</b>

- <em>bewegingswet</em>:
$$ s(\tau) = \left \{
	 \begin{array}{ll}
	  \hphantom{-}2\tau^2 & 0 \le \tau \le 0.5 \\
	  -2\tau^2 + 4\tau -1 & 0.5 \le \tau \le 1. \end{array} \right.
$$

<div style=" display: grid; place-items: center; place-content: center;">
 <img style="border: 0; display: inline-block;" height="100%"
      src="images/motion-law-bang-bang.png"
 >
</div>

- <em>continu&iuml;teit</em>: 2de orde. Ruk wordt oneindig bij $\tau=0$,
$\tau=0.5$ en $\tau=1$. De middelste discontinu&iuml;teit is dubbel zo groot als de beide andere.
- <em>voordeel</em>: deze bewegingswet heeft de <b>laagste piek-versnelling</b>, en vraagt de kleinste piek-stroom van de motor.
- <em>nadeel</em>: ook het vragen van een &ldquo;oneindige&rdquo; <em>ruk</em> resulteert nog altijd in hoge dynamische belasting (van motor <em>en</em> machine).


<b>Trapezoidale snelheid</b>

<div style=" display: grid; place-items: center; place-content: center;">
 <img style="border: 0; display: inline-block;" height="100%"
      src="images/motion-law-trapezoidal-velocity.png"
 >
</div>

- <em>bewegingswet</em>: variatie op <em>bang-bang versnelling</em>, met een stukje beweging aan continue snelheid toegevoegd in het midden van de beweging. Dit is nuttig als de gevraagde snelheid <em>beperkt</em> moet blijven.
- <em>continu&iuml;teit</em>: 2de orde. Er komt een extra &ldquo;inwendige&rdquo; oneindige ruk bij, maar de grootte van de discontinu&iuml;teiten is nu overal gelijk.


<b>Derde-orde polynoom</b>

- <em>bewegingswet</em>: $$ s(\tau) = 3\tau^2 - 2\tau^3.$$

<div style=" display: grid; place-items: center; place-content: center;">
 <img style="border: 0; display: inline-block;" height="100%"
      src="images/motion-law-3rd-order-polybomial.png"
 >
</div>

- <em>continu&iuml;teit</em>: 2de orde. Ruk wordt oneindig bij $\tau=0$ en $\tau=1$.
- <em>voordeel</em>: deze bewegingswet heeft de <b>laagste &ldquo;RMS&rdquo;-versnelling</b> (&ldquo;RMS&rdquo;, of <a href="https://nl.wikipedia.org/wiki/Kwadratisch_gemiddelde">kleinste kwadratisch gemiddelde</a>):
$$a_{rms} = \sqrt{ \int_0^1 a^2(\tau) \, d\tau}.$$
- <em>nadeel</em>: oneindige ruk, dus nog altijd hoge dynamische belasting (van motor <em>en</em> machine).


<b>Harmonische beweging</b>

- <em>bewegingswet</em>: $$ s(\tau) = \frac{1 - \cos{\pi\, \tau}}{2}.$$

- <em>continu&iuml;teit</em>: Ruk wordt oneindig bij $\tau=0$ en $\tau=1$.
- <em>voordeel</em>: continuiteit van de snelheid.
- <em>nadeel</em>: &ldquo;oneindige&rdquo; ruk resulteert in hoge dynamische belasting (van motor <em>en</em> machine).

<h3>Voorbeeld 1: servo-aandrijving</h3>

**Gegeven**:
- bewegingsinterval: $T = 0.2 s$.
- afstand tussen twee stilstanden: $L = 0.5m$.
- inertie van de last en de motor: $m = 20 kg$.
- bewegingswet: <em>bang-bang versnelling</em>.

**Gevraagd**:
- piek in de inerti&euml;le belasting.
- grootste discontinu&iuml;teit in de versnelling.

**Oplossing**:
- piekversnelling:
 $\frac{dS^2}{d\tau^2}|_\text{max} = \frac{L}{T^2} a_{\text{max}}
 = \frac{0.5}{0.2^2} \, {\color{blue}4} = 50 m/s^2$.
- piek in de inerti&euml;le belasting: $20 kg \, 50 \frac{m}{s^2} = 1000 N$.
- grootste discontinu&iuml;teit in de versnelling doet zich voor halverwege de beweging, op $t=0.1 s$, en is $2 \cdot 50 \frac{m}{s^2} = 100 \frac{m}{s^2}$.


<h3>Voorbeeld 2: nok/volger systeem</h3>

**Gegeven**:
- rotatiesnelheid van de nok: $n = 200\, rpm$, ofwel $\omega = 20.94\, rad/s$.
- bewegingsinterval van de nok: $\beta = \pi/2\, rad$.
- heffing tussen twee stilstanden: $L = 20\, mm = 0.02\, m$.
- bewegingswet: <em>cylcoide</em>.

**Gevraagd**:
- piek in de inerti&euml;le belasting.
- grootste discontinu&iuml;teit in de ruk.

**Oplossing**:
- piekversnelling:
 $\frac{dS^2}{dt^2}|_\text{max} = \frac{L}{\beta^2}\, \omega^2\, a_{\text{max}}
 = \frac{0.02}{\pi^2/4} \, 20.94^2\, {\color{blue}2\pi} = 22.3 m/s^2$.
- grootste discontinu&iuml;teit in de ruk doet zich voor bij $\theta =0$ en $\theta = \pi/2$. Dus:\
$\frac{dS^3}{dt^3}|_{\text{max}} = \frac{L}{\beta^3} \, \omega^3 \, j|_{\text{max}} = \frac{0.02}{\pi^3/8} \, 20.94^3 \, {\color{blue}38.48} = 1871\, m/s^3$.


<h3>Voorbeeld 3: synchronisatie van twee servo-aangedreven machine-assen</h3>

**Gegeven**:
- twee lineaire assen $(x,y)$ moeten een rechte lijn uitvoeren in het $xy$-vlak, met startpunt $(x_s,y_s)$ en eindpunt $(x_e,y_e)$. Dus met verplaatsingen $\Delta x = x_e -x_s$ en $\Delta y = y_e -y_s$.
- tijd beschikbaar voor de beweging is $T_m$.
- bemonsteringstijd van de controler is $T_s$.

**Gevraagd**:
- <em>interpolatie</em> van de beweging, d.w.z., een sequentie van tussenliggende posities, op elk bemonsteringstijdstip.

**Oplossing**:
- kies een <em>bewegingswet</em> $s(\tau)$.
- bemonsteren op $k\,T_s$ geeft:
$$ x(k\,T_s) = x_s + \Delta x \, s\left(\frac{k\, T_s}{T_m}\right),
\\
 y(k\,T_s) = y_s + \Delta y\, s\left(\frac{k\, T_s}{T_m}\right).$$

<h2>Periodische bewegingswetten</h2>

**Herhaling**:
- beweging herhaalt zich na een bepaalde tijd.
- de bewegingseigenschappen aan het eind zijn identiek aan die aan het begin.

**Aanpak**:
- stuksgewijze functies.
- Fourier-reeksen.

Deze functies zijn specifiek voor nokken ontworpen, en voldoen aan de &ldquo;<a href="#fundWet">fundamentele wet</a>&rdquo;: continu&iuml;teit in positie, snelheid en versnelling.

**Cycloides en half-cycloides**:
- versnelling is nul aan beide kanten.
- twee cycloides aan elkaar hangen houdt geen steek.

<h3>Fourier-reeksen</h3>

**Dimensieloze voorstelling** van de periodische bewegingswet, met $K$ harmonischen in de reeksontwikkeling:

$$
s(\tau) = \frac{a_0}{2} + \sum_{k=1}^{K} \left( a_k \, \cos(2\pi\, k\, \tau) + b_k \, \sin(2\pi\, k\, \tau) \right).
$$

Voor een nok is $\tau = \theta / 2\pi$, en voor een servo-aandrijving $\tau = t/T$.

**Voordelen** van Fourier-reeksen:
- alle afgeleiden zijn altijd continu.
- de harmonischen in de buurt van eigenfrekwenties kunnen uit de reeksontwikkeling van de bewegingswet verwijderd worden, zodat die frekwenties niet &ldquo;aangestoten&rdquo; worden.

**Nadeel**: heel moeilijk om een stilstand te realiseren. Daarvoor zou een &ldquo;oneindige&rdquo; reeks nodig zijn.
