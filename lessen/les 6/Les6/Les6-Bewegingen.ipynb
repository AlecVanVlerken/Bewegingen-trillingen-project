**Krachtsluiting**: er is een uitwendige kracht nodig om nok en volger in contact te houden gedurende de hele beweging. In de goedkoopste uitvoeringen komt die externe kracht van een combinatie van de zwaartekracht en een veer. Zwaartekracht dus niet vergeten en nok, ook mechanisme, goed positioneren.

**Vormsluiting**: de nok en volger houden contact omdat hun <em>geometrie</em> enkel &eacute;&eacute;n relatieve <em>bewegingsvrijheidsgraad</em> toelaat. - **groefnokken** (&ldquo;track&rdquo; of &ldquo;groove&rdquo; cams): er is één enkele volger, die door een groef geleid wordt om de nok te volgen. **Terminologie**:
- **traceerpunt** (&ldquo;trace point&rdquo;): voor een rolvolger is dat het centrum van de rol.
- **steekkromme** (&ldquo;pitch curve&rdquo;, &ldquo;theoretisch profiel&rdquo;): het pad van het traceerpunt ten opzichte van de nok.
- **basiscirkel** (&ldquo;base circle&rdquo;): kleinste ingeschreven cirkel rakend aan de *nok*, met centrum in nokrotatiecentrum. Met andere woorden, het nokprofiel wordt &ldquo;opgebouwd&rdquo; bovenop die basiscirkel.
- **steekcirkel** (&ldquo;pitch circle&rdquo;): idem, maar rakend aan de *steekkromme*.
- **drukhoek** (&ldquo;pressure angle&rdquo;): hoek tussen de richting van de ogenblikkelijke beweging van de volger, en de normaal op de steekcurve.
- **heffing** (&ldquo;stroke&rdquo;, &ldquo;throw&rdquo;, &ldquo;slaglengte&rdquo;): afstand tussen de twee <em>uiterste posities</em> van de volger. (En dus het verschil in straal van de steekcirkel en de basiscirkel.)

**Drukhoek $\alpha$**: deze hoek heeft een gelijkaardige functie als de *transmissiehoek* bij mechanismes:

$$ \text{drukhoek } = 90° - \text{transmissiehoek} $$

Voor $\alpha=0$ kan alle door de nok geleverde kracht gebruikt worden om de beweging van de volger aan te drijven.

Voor $\alpha \ne 0$ is er verlies aan aandrijfkracht omdat er *slip* optreedt.
<br>
(Tenzij in het geval van een volg-<em>roller</em>.)

Voor $\alpha = \pm 90°$ is er geen volgerbeweging meer. Of, correcter, er is geen <em>krachtoverdracht</em> meer van de ok naar de volger.

Dit zijn de krachten die inwerken op de volger (met $N$ de contactkracht in de normaalrichting):
- kracht in de richting van de volgerbeweging: $F = N\, \cos(\alpha)$.
- transversale kracht: $T = N\, \sin(\alpha)$.


Het overgedragen (en dus door wrijving verspilde) vermogen $P$ is:
$$P = T\, r \, \omega = F \, \tan(\alpha) \, r \, \omega.$$
(Wanneer de nok de volger raakt op een afstand $r$, en roteert met hoeksnelheid $\omega$.)

Die drukhoek is dus duidelijk al een eerste parameter om mee te nemen in jullie ontwerp-afwegingen.
want de transversale kracht veroorzaakt *wrijving* op de volger. Dit geeft aanleiding tot deze **vuistregels**:
- $|\alpha| \lt 30°$ voor een translatie-volger.
- $|\alpha| \lt 35°$ voor een rotatie-volger.

<h3>Analytisch ontwerp van een nok met een vlakke volger</h3>

(Dit is niet het type nok dat je voor je examen moet ontwerpen. Het is wel interessant om te zien of je de argumentaties achter het ontwerpverhaal kan volgen.)

**Gegeven**: hefwet $S=f(\theta)$, met $S$ de gewenste lineaire verplaatsing van de volger, en $\theta$ de nok-hoek.

**Gevraagd**: het vormprofiel van de nok, en de minimale lengte van het contactoppervlak van de vlakke volger.

**Aanpak**:
- genereer het nok-profiel als een curve met punten $x(\theta), y(\theta)$.
- controleer het profiel op **spitse punten** (&ldquo;cusps&rdquo;).
  + pas desgewenst het profiel aan.
- leid daarna de volger-lengte af.

<div style="height: 25rem;">
<img style="border: 0; display: inline-block;" height="100%"
      src="images/cam-profile.png"
>
</div>

**Nok-profiel**

Kies de straal $C$ van de basiscirkel. Dan is dit de afstand tussen het contactpunt nok/volger en het middelpunt van de nok: $R=C+f(\theta)$.

Druk $R$ en de positie van het contactpunt $l$ uit in functie van de nok-profiel-co&ouml;rdinaten $x$ en $y$:

\begin{aligned}
  R \quad &= \quad y\, \sin(\theta) + x\, \cos(\theta),
	\\
  l \quad &= \quad y\, \cos(\theta) - x\, \sin(\theta).
\end{aligned}

Het verband tussen deze beide vergelijkingen is dat $l = \frac{dR}{d\theta} = f'(\theta)$.

Los $x$ en $y$ nu op in functie van $R$ en $l$:

\begin{aligned}
  x \quad &= \quad (C + f(\theta)) \, \cos(\theta) - f'(\theta)\, \sin(\theta),
	\\
  y \quad &= \quad (C + f(\theta)) \, \sin(\theta) + f'(\theta)\, \cos(\theta).
\end{aligned}


<div style="height: 20rem;">
<img style="border: 0; display: inline-block;" height="100%"
      src="images/cam-profile-cusp.png"
>
</div>

**Controlleer op spitse punten** (&ldquo;cusps&rdquo;)

Een spits punt doet zich voor als $\frac{dx}{d\theta} = \frac{dy}{d\theta} = 0$. Want dan verandert het contactpunt niet wanneer de nok over een &ldquo;infinitesimale&rdquo; hoek $d\theta$ roteert. 
Met andere woorden, deze conditie

\begin{aligned}
  \frac{dx}{d\theta} \quad &= \quad -(C + f(\theta) + f''(\theta)) \, \sin(\theta) &= 0,
	\\
  \frac{dy}{d\theta} \quad &= \quad \hphantom{-}(C + f(\theta) + f''(\theta)) \, \cos(\theta)  &= 0,
\end{aligned}

kan vermeden worden als

$$ \forall \theta: \quad C + f(\theta) + f''(\theta) \gt 0,$$

want $\sin(\theta)$ en $\cos(\theta)$ kunnen nooit tegelijk nul worden.

**Pas nok-profiel aan**:
<br>
Als bovenstaande conditie niet is voldaan moet de basiscirkel vergroot worden. (In de veronderstelling dat de hefwet $f(\theta)$ niet mag gewijzigd worden.) En dan itereren we de bovenstaande oplossingsaanpak.

**Bereken volger-lengte**: de volgerlengte moet minimaal zo lang zijn als de maximale $l$ die zich over een volle nokrotatie kan voordoen. Dus:

$$ l_{\text{max}} = \max_{\theta} f'(\theta).$$


**Conclusie**: 
- voor de analyse is er niets meer &ldquo;complex&rdquo; nodig dan de eerste en tweede afgeleide van de hefwet $f(\theta)$.
- de beste oplossing volgt vaak pas na een aantal iteraties.

</div>

<h3><span style="color: Maroon;">Analytisch ontwerp van een nok met een volgrol</span></h3>
(Dit is het ontwerp dat jullie gaan maken voor het examen.)

**Gegeven**: 
- hefwet $S=f(\theta)$, met $S$ de gewenste lineaire verplaatsing van de volger, en $\theta$ de nok-hoek.
- straal $R_0$ van de steekcirkel.
- straal $R_r$ van de rol in de volger.

**Gevraagd**: het vormprofiel van de nok, d.w.z., vind de functie $R(\theta)$ van de <em>steekkromme</em>.

**Aanpak**: &emsp;(dit is enkel het overzicht; de details komen later)
- druk de steekkromme uit in polaire co&ouml;rdinaten. (Die passen op natuurlijke wijze op het feit dat de nok aangedreven wordt in rotatie.)
- controleer de grootte van de drukhoek $\alpha$.
- controleer voor **ondersnijding**.
- leid daarna het nok-profiel af: $x(\theta), y(\theta)$.

**Steekkromme in polaire co&ouml;rdinaten**: we weten dat die kromme de volgende analytische vorm heeft, $R(\theta) = R_0 + f(\theta)$, met $f(\theta)$ als de te vinden functie.

</div>

<div style="width: 86ch;">

**Drukhoek controleren**:

<div style="height: 30rem;">
<img style="border: 0; display: inline-block;" height="100%"
      src="images/cam-excentricity.svg"
>
</div>

De drukhoek $\alpha$ en excentriciteit $e$ zijn:
- $\alpha$ is de hoek tussen de contactnormaal en de as van de volger.
- $e$ is de afstand tussen de as van de volger en het centrum $O_1$ van de nok.

De contactkracht levert geen moment rond elk punt op de contactnormaal.
Het punt $O_2$ is de intersectie van die contactnormaal met de horizontale as.

De lijn $O_1 O_2$ staat loodrecht op de as van de volger,
dus al die punten hebben een zuiver *vertikale* snelheid, *parallel* aan die van de volger.

Van al deze punten is 
$O_2$ het enige wiens snelheid
$\dot{y} = |O_1 O_2| \omega$,
dezelfde *grootte* heeft as de snelheid van de volger:
de geometrie van de rode driehoek toont dat het *vermogen* geleverd door de contactkracht tegen de snelheid dezelfde is in het contactpunt
als in punt $O_2$, omdat zowel de kracht door beide punten, als de richting van de snelheid in beide punten, dezelfde zijn.

De snelheid $\dot{y}$ is evenredig met de hefsnelheid
$f'(\theta) = \frac{dy}{d\theta}$, met $y = f(\theta)$:
$$
\dot{y}
 = \frac{dy}{dt}
 = \frac{dy}{d\theta} \frac{d\theta}{dt}
 = f'(\theta) \omega.
$$

En omdat $\dot{y} = |O_1 O_2| \omega$, geldt de volgende gelijkheid:
$|O_1 O_2| = f'(\theta).$

Driehoeksformules in de rode en blauwe driehoeken geven:

$$
\tan(\alpha) = \frac{|O_1 O_2| - e}{a + y}, \quad \text{en } 
a = \sqrt{(R_b + R_r)^2 - e^2}.
$$
Waarbij $R_0 = R_b + R_r$, zodat de drukhoek $\alpha$ wordt:
$$
\alpha = \text{arctan} \left( \frac{f'(\theta) - e}{\sqrt{R_0^2 - e^2} + f(\theta)} \right).
$$

</div>


<div style="width: 76ch;">

Herinner: als je een kleinere drukhoek wil:
- verhoog je de straal van de steekkromme.
- verhoog je de excentriciteit. 

Een negatief gevolg van het laatste is wel dat de drukhoek *verhoogt* gedurende de terugkerende beweging want daar heeft de excentriciteit een ander teken&hellip; Hebben we dan eigenlijk iets gewonnen? Ja, <em>als</em> de toepassing tijdens die terugkeerbeweging een veel kleinere contactkracht nodig heeft.

</div>

**Controleer op ondersnijding**:

Als we $R_r$ vergroten, dan wordt $\rho_c$ kleiner.

Als we daarmee doorgaan tot $R_r = \rho$, dan is $\rho_c =0$.

Dus het nokprofiel wordt *spits* (zie figuur hieronder, links).

De rol-volger nog kleiner maken, resulteert in *ondersnijding* (zie figuur hieronder, rechts).

<div style="height: 20rem;">
<img style="border: 0; display: inline-block;" height="100%"
      src="images/cam-undercutting-2.png"

Dus, ondersnijding vermijd je door te zorgen dat:
$$\rho_{\text{min}} \gt R_r.$$

Dat doe je door:
- de straal $R_r$ van de rol-volger kleiner te maken.\
 Maar dat resulteert uiteindelijk in een te hoge contactdruk in de volger.
- de straal van de steekkromme te verhogen.\
 Dat kan &ldquo;onbeperkt&rdquo;, als de toepassing geen probleem heeft met hogere materiaalkosten en grotere inbouwgroottes.

Het is dus mogelijk dat je bovenstaande stappen een paar keer moet itereren, vooraleer je ontwerp aan alle specificaties voldoet. Er zijn opnieuw een aantal afwegingen te maken; bereid je daar op voor.

De <a href="https://nl.wikipedia.org/wiki/Kromtestraal#Parametrisch_beschreven_kromme">kromtestraal</a> van de *steekkromme* is:

$$ \rho =
 \frac{\left [ R^2 + (f'(\theta))^2 \right ]^{\frac{3}{2}}}
      {R^2 + 2(f'(\theta))^2 - R\, f''(\theta)},
\quad \text{met } R= R_0 +f(\theta).
$$


**Bereken het nok-profiel**.

De Cartesische co&ouml;rdinaten van een punt $C$ op de steekkromme zijn:

$$ x_C \quad = \quad R\, \cos(\theta), \\
   y_C \quad = \quad R\, \sin(\theta).
$$

Dus, de Cartesische co&ouml;rdinaten van een punt $A$ op het nok-profiel zijn:

$$ x_{\text{cam}} \quad = \quad R\, \cos(\theta) - R_r \, \cos(\theta + \alpha), \\
  y_{\text{cam}} \quad = \quad R\, \sin(\theta) - R_r\, \sin(\theta + \alpha).
$$

<h3>Dynamische analyse</h3>

De nok duwt de volger &ldquo;omhoog&rdquo;, en bijna altijd is er dan een *veer* om de volger tegen de nok te houden bij een &ldquo;neergaande&rdquo; beweging.

<div style="display: grid; grid-template-columns: 1fr 3fr; grid-column-gap: 1%;">
<div style="height: 20rem;">
<img style="border: 0; display: inline-block;" height="100%"
      src="images/cam-forces.png"
>
</div>
<div>

**Krachten** op het tarre lichaam van de volger:
- functionele krachten $F_{\text{func}}$.
- contactkracht $N$.
- veerkracht $F_v$.
- inerti&euml;le kracht.
- zwaartekracht.
- reactie- kracht $R$ en -moment $M$.
- wrijvingskracht $\mu R$.

</div>
</div>

**Vertikaal krachtevenwicht**:

$$
N\, \cos(\alpha) = F_{\text{func}} + F_v + m \frac{d^2y}{dt^2} + m\, g\, \cos(\gamma) + \mu R,
$$

met:

- $m$: de massa van het volger-lichaam.
- $\gamma$: hoek tussen de volgerbeweging en de vertikale.
- $\mu$: Coulomb-wrijvingsco&euml;fficient.
- $g$: zwaartekrachtversnelling.
- $y$: verplaatsing van de volger, positief in opwaartse richting.
- $\alpha$: drukhoek.


**Horizontaal krachtevenwicht**

Effect van de drukhoek $\alpha$:
- horizontaal evenwicht geeft: $|R| = |N\, \sin(\alpha)|$.
- dus de wrijvingskracht wordt groter als de drukhoek groter wordt.
- grote drukhoeken leiden tot *contactverlies* of *blokkage*.

**Resulterende contactkracht**

Indien:
- de volger een perfect star lichaam is.
- de glijwrijving in de volger verwaarloosbaar is.
- de rotatiesnelheid van de nok constant is.
- er permanent contact is tussen nok en volger, d.w.z. $y(t) := S(t) = S(\theta/\omega)$.

Dan is de contact-normaalkracht $N$ tussen nok en volger
$$
N= \frac{F_{\text{func}} + F_{v0} + k_v S + m \omega^2 \frac{d^2 S(\theta)}{d\theta^2} + mg \, \cos(\gamma)}{\cos(\alpha)},
$$

met $k_v$ de veerconstante, en $F_{v0}$ de voorspanning in die veer, by $S=0$.


**Dimensioneren van de veer**

Om <em>contact te garanderen</em> moet de normaalkracht $N$ altijd positief zijn, dus:

$$
F_{v0} + k_v S  \ge F_{\text{func}} - m \omega^2 \frac{d^2 S(\theta)}{d\theta^2} - mg \, \cos(\gamma).
$$

(Belangrijke vraag: vraagt de <em>toepassing</em> dat contact altijd <em>moet</em> gegarandeerd worden&hellip;?)

**&Eacute;&eacute;n** strategie is om <em>eerst</em> de voorspanning $F_{v0}$ te **kiezen**, en daaruit de veerconstante te berekenen:

$$
k_v  = \max_{\theta} \frac{- F_{v0} -F_{\text{func}} - m \omega^2 \frac{d^2 S(\theta)}{d\theta^2} - mg \, \cos(\gamma)}{S(\theta)}.
$$

De ontwerper moet de **afweging** maken tussen **voorspanning** en **veerconstante**:
- veel voorspanning geeft hoge contactkrachten op bepaalde nok-posities.
- geen voorspanning vereist een hoge veerconstante voor de locaties met kleine heffing $S$ en grote *inwaartse* versnelling.


**Vermogen doorgegeven tussen nok en volger**

Onder dezelfde &ldquo;ideale&rdquo; aannames en geen excentriciteit, geldt:

\begin{aligned}
P(\theta) &= N(\theta) \, \sin(\alpha) \, R(\theta) \omega
\\
 & = R(\theta) \omega \, \tan(\alpha) \left( F_{\text{func}} + F_{v0} + k_v S + m \omega^2 \frac{d^2 S(\theta)}{d\theta^2} + mg \, \cos(\gamma) \right).
\end{aligned}


Als de krachten vooral inerti&euml;el zijn (d.w.z., lage $F_{\text{func}}$ en lage wrijving), dan geldt volgende benadering, gezien vanuit de *volger*:

$$ P(\theta) \approx m \frac{d^2 S}{dt^2} \frac{dS}{dt}. $$
