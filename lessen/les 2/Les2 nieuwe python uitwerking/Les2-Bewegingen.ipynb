<h3>Dood punt &mdash; Singulariteit</h3>
Als het aandrijvend gelid 2 volledig roteert, dan is er geen gevaar dat het mechanisme blokkeert.
Maar als het aandrijvend gelid 2 <em>oscilleert</em> (zoals de <em>tuimelaar/rocker</em> in de figuur hieronder), dan zou het kunnen dat het mechanisme blokkeert op zijn uiterste standen,
omdat de <em>aangrijpingslijn</em> van de aandrijvende kracht ligt in de <em>richting van gelid 4</em>.
Dus, in dat geval moet er voldoende inertie (&ldquo;vliegwiel&rdquo;, &ldquo;flywheel&rdquo;) voorzien worden,
zodat het mechanisme <em>doorheen het dode punt</em> kan verderbewegen.

<h3>Overbrengings-/transmissie-hoek</h3>

<div style=" display: grid; place-items: center; place-content: center;">
 <div style="height: 15rem;">
 <img style="border: 0; display: inline-block;" height="100%"
      src="images/planar-four-bar-linkage.png"
 >
 </div>
</div>

De hoek $\gamma$ is de <em>overbrengingshoek</em>
tussen de <b>koppelstang</b> 3 en de <b>uitvoer</b> 4.
Uit de <a href="https://en.wikipedia.org/wiki/Law_of_cosines">cosinus-regel</a>, $c^2=a^2 + b^2 -2ab\cos(\gamma)$, volgt dat
$$
\gamma = \cos^{−1} \left[
\frac{
r_1^2 + r_2^2 − r_3^2 − r_4^2 − 2r_1 r_2 \cos{(\theta_2)}
}
{−2r_3 r_4}
\right].
$$

De overbrengingshoek ligt best zo dicht mogelijk bij een rechte hoek
(ver weg van de dode punten rond 0&deg; en 180&deg;), omdat dan het koppel van de aandrijving maximaal wordt doorgegeven aan de uitvoer. De figuur hieronder schetst twee transmissie-paden ("1" en "2") van hetzelfde invoerkoppel $\tau_i$: 
<ul>
<li>

de links zijn verbonden door rotatie-gewrichten, en die geven maar een deel van de kracht door naar het volgende gelid. Meer concreet, de krachten 1' en 2'.

<li>

aan het eind leveren die doorgegeven krachten opnieuw een uitvoer-koppel $\tau_o$.

<li>de configuratie waar de transmissie-hoek het dichtst bij 90 graden ligt, met name, de <b>zwarte</b> configuratie, is meer efficiënt in het doorgeven van het invoer-koppel naar de uitvoer dan de <span style="color:blue;"><b>blauwe</b></span> configuratie.
</ul>

<div style=" display: grid; place-items: center; place-content: center;">
 <div style="height: 15rem;">

      src="images/fourbar-linkage-inverted.png"

De <em>voorwaartse</em> positie-analyse start met een gegeven
$\theta_2$ (want daar zit doorgaans de aangedreven motor, waar een 
<a href="https://en.wikipedia.org/wiki/Rotary_encoder">hoek-encoder</a> die hoek $\theta_2$ meet),
en zoekt de hoeken $\theta_3,\theta_4$ (waar de &ldquo;taak&rdquo; van het mechanisme zit).

<em>Ad hoc</em> methode:

$\theta_4 = \pi - (\alpha + \beta)$, met
$\alpha = \cos^{-1}\left[ \frac{r_3^2 - z^2 -r_4^2}{-2 r_4 z}\right]$,
en $\beta = \cos^{-1}\left[ \frac{r_2^2 - z^2 -r_1^2}{-2 r_1 z}\right]$,

Dat geeft <em>twee oplossingen</em>, want:
$z^2 = r_1^2 +r_2^2 -2 r_1 r_2 \cos(\theta_2)$.

</div>

<h3>Positie-analyse gebaseerd op sluitingsvergelijkingen</h3>

<div style="height: 15rem;">
 <img style="border: 0; display: inline-block;" height="100%"
      src="images/planar-four-bar-linkage.png"
 >
</div>

<p><span style="color: Maroon;"><b>Sluitingsvergelijkingen</b></span>: we weten dat de vier stangen in het mechanisme altijd met elkaar verbonden zijn. Dus geldt de volgende gelijkheid:

$$\bf r_2 + r_3 = r_1 + r_4.$$

Projectie op de coordinaat-assen van het vaste gelid ($X$ horizontaal langs $r_1$, en $Y$ vertikaal) geeft de volgende twee scalaire vergelijkingen:

\begin{aligned}
X &: r_2 \cos(\theta_2) + r_3 \cos(\theta_3) -r_4 \cos(\theta_4) - r_1 = 0,
\\
Y &: r_2 \sin(\theta_2) + r_3 \sin(\theta_3) -r_4 \sin(\theta_4) = 0.
\end{aligned}

Dit geeft een stelsel van twee <b>niet-lineaire</b> vergelijkingen in de twee onbekenden $\theta_3$ en $\theta_4$.
Hiervoor bestaan <b>numerieke &ldquo;solvers&rdquo;</b> (bijv. gebaseerd op een <a href="https://en.wikipedia.org/wiki/Newton%27s_method">Newton-Raphson methode</a>), die starten met geschatte initi&euml;le waarden $\theta_3^0$ en $\theta_4^0$.

In veel software-programma's heet zo'n algemene &ldquo;solver&rdquo;
<tt>fsolve</tt>; dat is ook zo in de <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.fsolve.html">Python code</a> die we in dit vak gebruiken.

Conclusie: deze <em>numerieke</em> methode van de <em>sluitingsvergelijkingen</em> voor elke (onafhankelijke!) <em>kringloop</em> (&ldquo;loop&rdquo;) in het mechanisme is <b>heel generiek</b>. <span style="color: Maroon;"><b>We zullen die aanpak nog erg veel gebruiken in het vak</b></span>.

</div>

```python
# Animation of four bar

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from matplotlib.animation import FuncAnimation

%matplotlib widget
```

```python
# Example analysis of a four-bar mechanism.

# Data initialization (all data is in SI units)
# kinematic parameters (link lengths)
r1 = 0.1397 #m
r2 = 0.0508 #m
r3 = 0.1524 #m
r4 = 0.0762 #m
phi1 = 0 #rad

# parameters of actuation and simulation

# position analysis
phi3_init = 0  # initial condition for fsolve (phi3 and phi4)
phi4_init = 0.2  # determines the branch of your mechanism 

t_begin =  0     # start time of simulation
t_end   = 25     # end time of simulation
Ts      =  0.10  # time step of simulation
t = np.arange(t_begin, t_end + Ts, Ts)  # time vector

# initialization of driver
omega = 0.5      # driver frequency
A     = 1        # amplitude
# angular position, velocity and acceleration:
phi2   = 1 + A * np.sin(omega * t)
dphi2  = omega * A * np.cos(omega * t)
ddphi2 = -omega ** 2 * A * np.sin(omega * t)
```

```python
# Function to rotate a vector z over an angle theta:
def rotate_vector(z, theta):
    rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                                [np.sin(theta), np.cos(theta)]])
    return np.dot(rotation_matrix, z)
```

```python
# Define loop closure equations
def loop_closure_eqs(phi_init, phi2, r1, r2, r3, r4, phi1):
    phi3 = phi_init[0]
    phi4 = phi_init[1]

    # Loop closure equations
    F1 = r2 * np.cos(phi2) + r3 * np.cos(phi3) - r4 * np.cos(phi4) - r1 * np.cos(phi1)
    F2 = r2 * np.sin(phi2) + r3 * np.sin(phi3) - r4 * np.sin(phi4) - r1 * np.sin(phi1)

    return [F1, F2]
```

```python
# Define kinematic computations of the four-bar mechanism

def kinematics_4bar(r1, r2, r3, r4, phi1, phi2, dphi2, ddphi2, phi3_init, phi4_init, t):
    global phi3, phi4, dphi3, dphi4, ddphi3, ddphi4, cond
    phi3   = np.zeros_like(t)
    phi4   = np.zeros_like(t)
    dphi3  = np.zeros_like(t)
    dphi4  = np.zeros_like(t)
    ddphi3 = np.zeros_like(t)
    ddphi4 = np.zeros_like(t)

    cond   = np.zeros_like(t)

    optim_options = {"full_output":True}  # options for fsolve

    for k, time in enumerate(t):
        # Position Analysis
        x, _, ier, message  = fsolve(lambda x: loop_closure_eqs(x, phi2[k], r1, r2, r3, r4, phi1), [phi3_init, phi4_init], **optim_options)

        if ier != 1:
            print("The fsolve exit flag was not 1, probably no convergence!")
            print(message)

        phi3[k] = x[0]
        phi4[k] = x[1]

        # Velocity Analysis
        A = np.array([[-r3 * np.sin(phi3[k]), r4 * np.sin(phi4[k])],
                      [r3 * np.cos(phi3[k]), -r4 * np.cos(phi4[k])]])
        B = np.array([r2 * np.sin(phi2[k]) * dphi2[k],
                      -r2 * np.cos(phi2[k]) * dphi2[k]])

        x = np.linalg.solve(A, B)
        dphi3[k] = x[0]
        dphi4[k] = x[1]
        cond[k]  = np.linalg.cond(A)

        # Acceleration Analysis
        B = np.array([r2 * np.cos(phi2[k]) * dphi2[k]**2 + r2 * np.sin(phi2[k]) * ddphi2[k] + r3 * np.cos(phi3[k]) * dphi3[k]**2 - r4 * np.cos(phi4[k]) * dphi4[k]**2,
                      r2 * np.sin(phi2[k]) * dphi2[k]**2 - r2 * np.cos(phi2[k]) * ddphi2[k] + r3 * np.sin(phi3[k]) * dphi3[k]**2 - r4 * np.sin(phi4[k]) * dphi4[k]**2])

        x = np.linalg.solve(A, B)
        ddphi3[k] = x[0]
        ddphi4[k] = x[1]

        # Next iteration step
        phi3_init = phi3[k] + (t[1] - t[0]) * dphi3[k]
        phi4_init = phi4[k] + (t[1] - t[0]) * dphi4[k]
```

```python
# Animation of mechanism motion
global fig, ax, x_left, x_right, y_bottom, y_top, P, S, index_vec
# compute width and height of drawing canvas:
x_left   = -1.25 * r2
y_bottom = -1.25 * max(r2, r4)
x_right  = r1 + 1.25 * r4
y_top    = 1.25 * max(r2, r4)

# Don't show figure until drawn explicitly:
plt.ioff()
# Initialize the figure
fig, ax = plt.subplots()

# Define the function to update the plot for each frame:
t_size = len(t)
# fraction of simulated frames in the animation:
frames = t_size/5

# Define point P and S as ground link:
P = np.array([0,0])
S = rotate_vector(np.array([r1, 0]), phi1)

# time between frames
delta = int(np.floor(t_size / frames))
# index vector for frames        
index_vec = np.arange(0, t_size, delta)

# calculation and visualisation of the kinematics:
kinematics_4bar(r1, r2, r3, r4, phi1, phi2, dphi2, ddphi2, phi3_init, phi4_init, t)

# next step in the animation:
def update(frame):
    ax.cla()  # Clear the current axes
    ax.axis('equal')  # Ensures 1 unit on x == 1 unit on y
    ax.set_xlabel('[m]')
    ax.set_ylabel('[m]')
    ax.set_xlim([x_left, x_right])
    ax.set_title(f'Frame {frame}')

    # Calculate Cartesian coordinates using rotation matrix
    Q = P + rotate_vector(np.array([r2, 0]), phi2[frame])
    R1 = Q + rotate_vector(np.array([r3, 0]), phi3[frame])
    R2 = S + rotate_vector(np.array([r4, 0]), phi4[frame])
    loop1 = np.array([P, Q, R1, R2, S])
    ax.plot(loop1[:, 0], loop1[:, 1], '-o')
    return(ax)

# Update the animation
ani = FuncAnimation(fig, update, frames=len(index_vec), interval=50)
```

De tijdsafgeleide van de <em>positie</em>-sluitingsvergelijkingen 

\begin{aligned}
&X: r_2 \cos(\theta_2) + r_3 \cos(\theta_3) -r_4 \cos(\theta_4) - r_1 = 0,
\\
&Y: r_2 \sin(\theta_2) + r_3 \sin(\theta_3) -r_4 \sin(\theta_4) = 0.
\end{aligned}

geeft 

\begin{aligned}
&X: −r_2 \sin(\theta_2) \omega_2 − r_3 \sin(\theta_3) \omega_3 + r_4 \sin(\theta_4)  \omega_4 = 0,
\\
&Y: r_2 \cos(\theta_2) \omega_2 + r_3 \cos(\theta_3) \omega_3 − r_4 cos(\theta_4 ) \omega_4 = 0.
\end{aligned}

Met andere woorden, een <b>lineair</b> stelsel van <b>twee onbekenden</b> $\omega_3$ en $\omega_4$.

<h3>Versnellings-analyse met geometrisch inzicht</h3>

Heel erg gelijkaardig aan de snelheidsanalyse:
gegeven $\theta_2$, $\theta_3,\theta_4$, $\omega_2$, $\omega_3$, $\omega_4$, en $\alpha_2$ (de &ldquo;gemeten&rdquo;) hoekversnelling van gewricht 2), vind de hoekversnellingen $\alpha_3$ en $\alpha_4$. We starten met het inzicht dat we de lineaire versnellingen van de gewrichten A en B als volgt kunnen linken met de hoeksnelheden en -versnellingen:

\begin{aligned}
a_A &= \alpha_2 \times r_2 + \omega_2 \times \omega_2 \times r_2, \quad \text{(gekend!)}
\\
a_B &= \alpha_4 \times r_4 + \omega_4 \times \omega_4 \times r_4,
\\
a_B &= a_A + \alpha_3 \times r_3 + \omega_3 \times \omega_3 \times r_3.
\end{aligned}

Het resultaat is opnieuw een stelsel van <b>lineaire</b> vergelijkingen in de onbekenden $\alpha_3$ en $\alpha_4$.

</div>

<h3>Versnellings-analyse gebaseerd op sluitingsvergelijkingen</h3>

De tijdsafgeleide van de snelheids-sluitingsvergelijkingen 

\begin{aligned}
 0 =&−r_2 \sin(\theta_2) \omega_2 − r_3 \sin(\theta_3) \omega_3 + r_4 \sin(\theta_4)  \omega_4,
\\
 0 =&r_2 \cos(\theta_2) \omega_2 + r_3 \cos(\theta_3) \omega_3 − r_4 cos(\theta_4 ) \omega_4.
\end{aligned}

geeft

\begin{aligned}
0 = &−r_2 cos(θ_2)ω_2^2 − r_3 cos(θ_3)ω_3^2 + r_4 cos(θ_4)ω_4^2,
\\
 &− r_2 sin(θ_2)α_2 − r_3 sin(θ_3)α_3 + r_4 sin(θ_4)α_4,
\\
0 = &−r_2 sin(θ_2)ω_2^2 − r_3 sin(θ_3)ω_3^2 + r_4 sin(θ_4)ω_4^2
\\
 &+ r_2 cos(θ_2)α_2 + r_3 cos(θ_3)α_3 − r_4 cos(θ_4)α_4.
\end{aligned}

Het resultaat is opnieuw een stelsel van <b>lineaire</b> vergelijkingen in de onbekenden $\alpha_3$ en $\alpha_4$.

<h3>Positie-analyse gebaseerd op sluitingsvergelijkingen</h3>

<div style="height: 14rem;">
 <img style="border: 0; display: inline-block;" height="100%"
      src="images/kruk-drijfstang.png"
 >
</div>

<b>Gegeven</b> $\theta$, <b>vind</b> $\phi$ en $x$.

<p id="crank-position-closure"><b>Sluitingsvergelijking</b>:

\begin{aligned}
X:  0 &= R \cos(\theta) + L \cos(\phi) - (L+R-x),
\\
Y:  0 &= R \sin(\theta) - L \sin(\phi).
\end{aligned}

Dit levert een stelsel <b>niet-lineaire</b> vergelijkingen op, in de twee onbekenden $\phi$ en $x$.
Dit stelsel kan <b>numeriek</b> opgelost worden, met beginschattingen van de onbekenden als input.


<div style="width: 76ch;">

<h3>Benaderende positie-analyse</h3>

Aangezien vaak $L \approx 4R$, kan <a href="#SE">deze positie-vergelijking</a> benaderd worden als

$$x \approx
R(1 − \cos \omega t) +
\frac{R^2}{4L}
(1 − \cos 2\omega t).
$$

Met gelijkaardige resultaten voor snelheid en versnelling:

\begin{aligned}
v &\approx R \omega \sin(\omega t) + \frac{R^2}{2L} \sin(2 \omega t),
\\
a &\approx R \omega^2 \cos(\omega t) + \frac{R^2}{L} \cos(2 \omega t),
\end{aligned}

Deze vergelijkingen leveren een goede basis voor de studie van <b>trillingen</b> in machines. Merk ook op dat de tweede harmonische <b>belangrijker</b> wordt in hogere afgeleiden.
